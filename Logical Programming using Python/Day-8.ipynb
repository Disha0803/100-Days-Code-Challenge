#### Binary Serach Tree

```python
class BST:
    def __init__(self,key):
        self.key=key
        self.lchild=None
        self.rchild=None
    def insert(self,data):
        if self.key is None:
            self.key=data
            return
        if self.key==data:
            print("Duplicate values not allowed")
            return
        if self.key>data:
            if self.lchild:
                self.lchild.insert(data)
            else:
                self.lchild=BST(data)
        else:
            if self.rchild:
                self.rchild.insert(data)
            else:
                self.rchild=BST(data)
    def search(self,data):
        if self.key==data:
            print("Node Found")
            return
        if data<self.key:
            if self.lchild:
                self.lchild.search(data)
            else:
                print("Node is not present")
        else:
            if self.rchild:
                self.rchild.search(data)
            else:
                print("Node is not present")

    def preorder(self):
        print(self.key,end=" ")
        if self.lchild:
            self.lchild.preorder()
        if self.rchild:
            self.rchild.preorder()

    def inorder(self):
        if self.lchild:
            self.lchild.inorder()
        print(self.key,end=" ")
        if self.rchild:
            self.rchild.inorder()

    def postorder(self):
        if self.lchild:
            self.lchild.postorder()
        if self.rchild:
            self.rchild.postorder()
        print(self.key,end=" ")

    def delete(self,data,curr):
        if self.key is None:
            print("Empty Tree")
            return
        if data<self.key:
            if self.lchild:
                self.lchild=self.lchild.delete(data,curr)
            else:
                print("Node absent")
        elif data>self.key:
            if self.rchild:
                self.rchild=self.rchild.delete(data,curr)
            else:
                print("Node absent")
        else:
            if self.lchild is None:
                temp=self.rchild
                if data==curr:
                    self.key=temp.key
                    self.lchild=temp.lchild
                    self.rchild=temp.rchild
                    temp=None
                    return
                self=None
                return temp
            if self.rchild is None:
                temp=self.lchild
                if data==curr:
                    self.key=temp.key
                    self.lchild=temp.lchild
                    self.rchild=temp.rchild
                    temp=None
                    return
                self=None
                return temp
            node=self.rchild
            while node.lchild:
                node=node.lchild
            self.key=node.key
            self.rchild=self.rchild.delete(node.key,curr)
        return

    def min_node(self):
        current=self
        while current.lchild:
            current=current.lchild
        print("\nMinimum node =",current.key)
    def max_node(self):
        current=self
        while current.rchild:
            current=current.rchild
        print("Maximum node =",current.key)

def count(node):
    if node is None:
        return 0
    else:
        return 1+count(node.lchild)+count(node.rchild)


root_value=input("Enter a root value to create a tree: ")
root=BST(root_value)
ch=0
while ch!=8:
    print("\nDo you want to continue?")
    print("1--> Insert into Tree\n2--> Search\n3--> Traverse the tree\n4--> Delete\n5--> Minimum Node\n6--> Maximum Node\n7--> Total Number of Nodes\n8--> Exit")
    ch=int(input())
    if ch==1:
        n=int(input("How many items you want to enter? "))
        print("Enter the items:")
        l=[]
        for i in range(n):
            l.append(input())
        for i in l:
            root.insert(i)
    elif ch==2:
        data=input("Enter the node you want to search: ")
        root.search(data)
    elif ch==3:
        c=input("In which order you want to display?\npre--> Preorder\nin--> Inorder\npost--> Postorder\n")
        if c=='pre':
            root.preorder()
        elif c=='in':
            root.inorder()
        elif c=='post':
            root.postorder()
        else:
            print("Invalid Input! Showing the tree by default in preorder")
            root.preorder()
    elif ch==4:
        data=input("Enter a node you want to delete: ")
        if count(root)>1:
            root.delete(data,root.key)
        else:
            print("Can't perform deletion operation beacuse of only a single root node")
    elif ch==5:
        root.min_node()
    elif ch==6:
        root.max_node()
    elif ch==7:
        print("Total number of nodes: ",count(root))
    elif ch==8:
        print("Thank You")
        exit()
    else:
        print("Enter a valid choice")
```
